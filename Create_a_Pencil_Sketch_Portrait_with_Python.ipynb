**Importing the necessary libraries**

```python
import cv2
from matplotlib import pyplot as pyplot
```

**Read Image**

```python
image= cv2.imread("chatrapati-shivaji-maharaj.png (1).png")
```

**Display the image**

```python
image= cv2.imshow(image)
```

**Read the image in RGB format**

```python
from matplotlib import pyplot as plt
%matplotlib  inline
img_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.show()
```

**Create new image by converting the original image to greyscale**

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gray_image
```

```python
plt.imshow(gray_image)
```

**Invert the new grayscale image**

```python
inverted_image = 255 - gray_image
plt.imshow(inverted_image)
```

**Blur the image by using the Gaussian Function**


```python
blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)
plt.imshow(blurred)
```

**Invert the blurred image into a pencil sketch**

```python
inverted_blurred = 255 - blurred
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale=268.0)
plt.imshow( pencil_sketch)
cv2.waitKey(0)
```
